```python
np.random.seed(10)
a=np.random.randint(10,200,50)
a
```

create a new array<br>
a>5=0 and a<=150 -good<br>
a<=50 -bad<br>
a>=150-excellent<br>

```python
def number_cat(n):
    if n<50:
        return 'bad'
    elif (n>=50) and (n<=150):
        return 'good'
    else:
        return 'excellent'
```

```python
number_cat_vect=np.vectorize(number_cat)
number_cat_vect(a)
```

```python
import pandas as pd
cat=pd.Series(number_cat_vect(a))
cat
```

```python
cat.value_counts()
```

```python
y_value=cat.value_counts()
x_value=cat.value_counts().index
plt.bar(x_value,y_value)
```

```python
plt.bar(x_value,y_value,color='grb')
colors={'good':'green','bad':'red','excellent':'blue'}
labels=list(colors.keys())
handles=[plt.Rectangle((0,0),1,1,color=colors[label])for label in labels]
plt.legend(handles,labels)
```
