```python
segments=data.Customer_Segment.value_counts().index
segments
```

```python
Sales_Segments=[]
for segment in segments:
    Sales_Segments.append((data['Sales'][data.Customer_Segment==segment]).sum())
Sales_Segments
```

```python
#create a pandas series with values asa sum in each
#category and index as segment name
Sales_Segments=pd.Series(Sales_Segments,segments)
Sales_Segments
```

```python
#Findout the segment name where we have the highest sales
Sales_Segments.idxmax()
```

```python
def cat_to_num(x):
    if x=='NOT SPECIFIED':
        return 0
    elif x=='LOW':
        return 1
    elif x=='MEDIUM':
        return 2
    elif x=='HIGH':
        return 3
    elif x=='CRITICAL':
        return 4

    
    
```

```python
data['Order_Priority_Number']=data['Order_Priority'].apply(cat_to_num)
data['Order_Priority_Number'].value_counts()
```
