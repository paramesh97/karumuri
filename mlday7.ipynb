```python
names=['ram','raj','ravi','roja','teja','raveendra','harish']
#find out the names whose have character 'e'
for i in names:
    if re.search('e',i):
        print(i)
```

```python
#find the names which starts with 'ra' and have atleast 4 chars
for i in names:
    if re.match('ra..',i):
        print(i)
    
```

```python
#find the names whose length is greater than are equal to 4
for i in names:
    if re.match('....',i):
        print(i)
    
```
