```python
help(make_regression)
```

```python
x,y=make_regression(100,1,n_targets=1,random_state=34)#targets=how many number of outputs,1=no of columns
print(x)
print(y)
```

```python
print(y.shape)
print(type(y))
print(y.ndim)
```

```python
plt.plot(x,y,'o')
```

```python
#import the linear regression model
from sklearn.linear_model import LinearRegression
#create an instance
LR=LinearRegression()
#Train
LR.fit(x,y)
#predict with train data
y_pred=LR.predict(x)
```

```python
print(LR.coef_)
print(LR.intercept_)
```

```python
#create user defined function with name mse
#inputs y_pred,y
#output is mean squared error
```

```python
def mse(y,y_pred):
    return((y_pred-y)**2).mean()
print(mse(y,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y,y_pred))
```
